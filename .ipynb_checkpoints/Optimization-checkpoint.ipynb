#Optimization!

```julia
# set new ICs and remember the old
real_xi = [x for x in sim.xi]
real_yi = [y for y in sim.yi]
real_zi = [z for z in sim.zi]
real_x0 = [p._x for p in sim.w.particles]
real_y0 = [p._y for p in sim.w.particles]
real_z0 = [p._z for p in sim.w.particles]

for i in 1:sim.w.n
    sim.xi[i] = real_x0[i]
    sim.yi[i] = real_y0[i]
    sim.zi[i] = real_z0[i]
end
opt = nb.Optimization(sim);
```

```julia
for i in 1:sim.w.n
    opt.sim.xi[i] = real_xi[i]
    opt.sim.yi[i] = real_yi[i]
    opt.sim.zi[i] = real_zi[i]
end

# fast gradient...
nb.grad!(opt)

fg = Float64[]
for i in 1:500
    push!(fg, opt.gx[i])
end

# direct gradient...
g0 = nb.grade(opt)
dg = Float64[]
D = sqrt(opt.sim.w.smth2)./200000.0
for i in 1:500
    print(i,".")
    opt.sim.xi[i] += D
    nb.exec!(sim, false, silent=true)
    opt.sim.xi[i] -= D
    fw = (nb.grade(opt)-g0)/D*opt.sim.w.n

    push!(dg, fw)
end
```

```julia
scatter(dg[dg.<10], fg[dg.<10], 0.1)
```

```julia
linreg(dg, fg)
```

```julia
nb.plot(opt)
```

```julia
step=0
N = 2000
ga = Float64[]
Winston.set_default_plot_size(700,700)
import IJulia
for l_target_t in linspace(log(opt.sim.tf*0.99), log(opt.sim.ti), N)    
    step+=1
    target_t = exp(l_target_t)
    
    println("----------------------- ti=",target_t," comp: ",step/N*100)
    sim.ti = target_t
    g=nb.optimize(opt, 3, 500.0)
    send_ipython(IJulia.publish, msg_pub(IJulia.execute_msg, "clear_output", [:wait => true]))
    #? nb.exec!(sim, false, silent=true)
    display(nb.plot(opt))
    push!(ga, g/500.0)
    display(plot([1:length(ga)], ga))
    
    #pp = FramedPlot()
    #add(pp, Histogram(hist(opt.grad[opt.grad.>1000])...))    
    #display(pp)
    println("==============================================")
    @show maximum(opt.grad)
    println("g=",g/500.0)
    if g/400000.0>1.0
        break
    end
    #break
end
```
