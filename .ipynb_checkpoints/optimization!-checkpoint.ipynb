```julia
N = 1048
stepc = 120
sim = nb.Simulation(nb.worldspherical(N, smth=0.4, opening_alpha=0.5, scale=5.7, xy=false),
limit_by_steps=true, stepc=stepc, n_test_particle=3*N);

nb.exec!(sim, false, silent=true)
display(nb.plot(sim))

real_xi = deepcopy(sim.xi)
real_yi = deepcopy(sim.yi)
real_zi = deepcopy(sim.zi)

real_xf = [p._x for p in sim.w.particles]
real_yf = [p._y for p in sim.w.particles]
real_zf = [p._z for p in sim.w.particles]

for i in 1:sim.w.n
    sim.xi[i] = real_xf[i]
    sim.yi[i] = real_yf[i]
    sim.zi[i] = real_zf[i]
end

opt = nb.Optimization(sim);
```

```julia
for i in 1:24
    println("------------------------------")
    @show i*5
    sim.stepf=i*5
    if i<16
        nb.optimize(opt, sim, 10, 0.5)
    else
        nb.optimize(opt, sim, 17, 1.0)
    end
    nb.exec!(sim, false, silent=true)
    @show nb.grade(opt, sim)
end
```

```julia
nb.grade(opt,sim)
```

#Now Cosmological

```julia
addprocs(8)
```

```julia
push!(LOAD_PATH,"./src/")
@everywhere begin
    import nbody
    reload("nbody")
    nb = nbody;
end
```

```julia
workers()
```

```julia
N = 2048
stepc = 120
sim = nb.Simulation(nb.worldspherical(N, smth=0.2, opening_alpha=0.5, space=nb.Cosmological, scale=9.1),
limit_by_steps=false, stepc=36, ti=0.02, tf=1.0, n_test_particle=3*N);
```

```julia
nb.exec!(sim, false, silent=false)
display(nb.plot(sim))
```

```julia
real_xi = deepcopy(sim.xi)
real_yi = deepcopy(sim.yi)
real_zi = deepcopy(sim.zi)

real_xf = [p._x for p in sim.w.particles]
real_yf = [p._y for p in sim.w.particles]
real_zf = [p._z for p in sim.w.particles]

for i in 1:sim.w.n
    sim.xi[i] = real_xf[i]
    sim.yi[i] = real_yf[i]
    sim.zi[i] = real_zf[i]
end

opt = nb.Optimization(sim);
```

```julia
for lti in linspace(log(0.97),log(0.02), 50)
    ti = exp(lti)
    println("------------------------------")
    @show ti
    sim.ti = ti
    nb.optimize(opt, sim, 17, 1.0)
    nb.exec!(sim, false, silent=true)
    @show nb.grade(opt, sim)
end
```

```julia
using Winston
Winston.set_default_plot_size(700,700)
hold(false)
scatter(sim.xi, sim.yi, 0.1, color="red")
hold(true)
display(scatter(real_xi, real_yi, 0.1, color="green"))
x=Float64[]
y=Float64[]
d=Float64[]
for i in 1:sim.w.n
    push!(x, real_xi[i])
    push!(x, sim.xi[i])
    dx = real_xi[i] - sim.xi[i]
    push!(x, NaN)
    push!(y, real_yi[i])
    push!(y, sim.yi[i])
    dy = real_yi[i] - sim.yi[i]
    push!(y, NaN)
    push!(d, sqrt(dx*dx+dy*dy))
end
plot(x,y)
```

```julia
p = FramedPlot(
         title="Title",
xlabel="Distance from real initial position",
ylabel="Number of particles")

add(p, Histogram(hist(d, 100)...))
```
